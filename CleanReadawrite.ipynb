```python
import pandas as pd
import numpy as np
import string
import re

df = pd.read_csv(r'C:\Users\paton\Pick2ReadData\Data\readawrite1.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.index
```

```python
df.dropna(subset=['title', 'category','author'],inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop_duplicates()

wordlist = '&#039;|&amp;|&bull;|&hearts;|&quot;|&lt;'
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())

df['title'] = df['title'].str.replace(wordlist, '')
df['author'] = df['author'].str.replace(wordlist, '')
df['category'] = df['category'].str.replace(wordlist, '')


#df[['title','author']] = df[['title','author']].replace(wordlist, '')
df
```

```python
#delete common word
df2 = df
commonword = ['นิยายแปล','เรื่องสั้น','รวมฉาก','อ่านฟรี','!','จบ','จบแล้วจ้า','ตอนพิเศษ','ทดลองอ่าน','18+']
wordcut = '|'.join(commonword)
df2['title'] = df2['title'].str.replace(wordcut,'')
df2
```

```python
df3 = pd.DataFrame(df2.title.str.split(r'#.*?(?=\S|$)',1).tolist(), 
                         columns = ['new_title','hashtag'])
df3['new_title'] = df3['new_title'].str.replace(' ', '')
df3['hashtag'] = df3['hashtag'].str.replace(' ', '')

df3['new_title'] = df3['new_title'].str.replace('.', '')
df3['hashtag'] = df3['hashtag'].str.replace('.', '')

#df[['new_title','hashtag']] = df.title.str.split(r'#.*?(?=\s|$)',expand = True) 

df3['hashtag'] = df3['hashtag'].str.extract(r'([\ก-\๙]+)')
df3['new_title'] = df3['new_title'].str.extract(r'([\ก-\๙]+)')


df3[df3['hashtag'].str.contains('') == True]
```
