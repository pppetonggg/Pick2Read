```python
import pandas as pd
import numpy as np
import string
import re

#df = pd.read_csv(r'C:\Users\paton\Pick2ReadData\Data\readawrite1.csv')
#df = pd.read_csv(r'C:\Users\paton\Pick2ReadData\Data\readawrite_fiction_AL.csv')
#too large when push on github so I move it in C drive
df = pd.read_csv(r'C:\Users\paton\readawrite_fiction_2.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
#print(df8.index)
print('---------------')
print(df.index)
#df = df.head(11)
df
```

```python
df.dropna(subset=['title', 'category','author'],inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop_duplicates()

wordlist = '&#039;|&amp;|&bull;|&hearts;|&quot;|&lt;'
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())

df['title'] = df['title'].str.replace(wordlist, '')
df['author'] = df['author'].str.replace(wordlist, '')
df['category'] = df['category'].str.replace(wordlist, '')

#df[['title','author']] = df[['title','author']].replace(wordlist, '')
df
```

```python
df = df[df['category'].str.contains('ผจญภัย') == True]
df.reset_index(drop=True, inplace=True)
df['new_title'] = df['title'].str.replace(' ', '')
df['new_title'] = df['title'].str.replace('.', '') 
df['new_title'] = df['title'].str.extract(r'([\ก-\๙]+)')
df.to_csv('readawrite_action.csv')
df
```

```python
#delete common word
df2 = df
commonword = ['นิยายแปล','เรื่องสั้น','รวมฉาก','อ่านฟรี','!','จบ','จบแล้วจ้า','ตอนพิเศษ','ทดลองอ่าน','18+']
wordcut = '|'.join(commonword)
df2['title'] = df2['title'].str.replace(wordcut,'')
df2
```
