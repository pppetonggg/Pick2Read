```python
df['ratingNoTweet_q'] = df['ratingNoTweet_q'].fillna(0)
df['ratingHashReview_q'] = df['ratingHashReview_q'].fillna(0)
df[['title','countTweet','ratingNoTweet_q','ratingHashReview_q']].sort_values(by = 'countTweet', ascending = False)
```

```python
#rating of website
df['ratingCount'] = df['ratingCount'].str.replace(',', '')
df['viewer'] = df['viewer'].str.replace(',', '')
df['favouriteCount'] = df['favouriteCount'].str.replace(',', '')
df[['ratingCount', 'viewer','favouriteCount']] = df[['ratingCount', 'viewer','favouriteCount']].apply(pd.to_numeric)

df['ratingLike_q'] = pd.qcut(df['ratingCount'], 5, labels=[1,2,3,4,5])
df['ratingView_q'] = pd.qcut(df['viewer'], 5, labels=[1,2,3,4,5])
df['ratingAdd2Fav_q'] = pd.qcut(df['favouriteCount'], 5, labels=[1,2,3,4,5])

df[['ratingView_q','ratingLike_q','ratingAdd2Fav_q']] = df[['ratingView_q','ratingLike_q','ratingAdd2Fav_q']].astype('float64')

#df['ratingWebsite'] = (((df.ratingView_q*2)+(df.ratingLike_q))/3).round(1)
df['ratingWebsite'] = ((df.ratingView_q+df.ratingLike_q+df.ratingAdd2Fav_q)/3).round(1)

#check
print(df['ratingLike_q'].unique())
print(df['ratingView_q'].unique())
print(df['ratingAdd2Fav_q'].unique())

df[['title','viewer','ratingCount','ratingView_q','ratingLike_q','ratingAdd2Fav_q','ratingWebsite']] 
```

```python
#ver 1
top = 96.0
w_hReview = 78/top
w_view = 96/top
w_like = 74/top
w_NoTweet = 70/top
w_add2Fav = 88/top

df['finalRating1'] = ((df.ratingLike_q+df.ratingView_q+df.ratingAdd2Fav_q+df.ratingNoTweet_q+df.ratingHashReview_q)/5).round(1)
df['finalRating1'].values[df['finalRating1'].values < 1.0] = 1.0

df['finalRating1_w'] = (((df.ratingLike_q*w_like)+(df.ratingView_q*w_view)+(df.ratingAdd2Fav_q*w_add2Fav)+(df.ratingNoTweet_q*w_NoTweet)+(df.ratingHashReview_q*w_hReview))/(w_view+w_like+w_add2Fav+w_NoTweet+w_hReview)).round(1)
df['finalRating1_w'].values[df['finalRating1_w'].values < 1.0] = 1.0

df['finalRating2'] = ((df.ratingLike_q+df.ratingView_q+df.ratingAdd2Fav_q+df.ratingNoTweet_q)/4).round(1)
df['finalRating2'].values[df['finalRating2'].values < 1.0] = 1.0

df['finalRating2_w'] = (((df.ratingLike_q*w_like)+(df.ratingView_q*w_view)+(df.ratingAdd2Fav_q*w_add2Fav)+(df.ratingNoTweet_q*w_NoTweet))/(w_view+w_like+w_add2Fav+w_NoTweet)).round(1)
df['finalRating2_w'].values[df['finalRating2_w'].values < 1.0] = 1.0

df['finalRating3'] = ((df.ratingLike_q+df.ratingView_q+df.ratingAdd2Fav_q)/3).round(1)
df['finalRating3'].values[df['finalRating3'].values < 1.0] = 1.0

df['finalRating3_w'] = (((df.ratingLike_q*w_like)+(df.ratingView_q*w_view)+(df.ratingAdd2Fav_q*w_add2Fav))/(w_view+w_like+w_add2Fav)).round(1)
df['finalRating3_w'].values[df['finalRating3_w'].values < 1.0] = 1.0

df[['title','viewer','ratingCount','favouriteCount','hashtag','ratingAdd2Fav_q','ratingView_q','ratingLike_q','ratingNoTweet_q','finalRating1','finalRating1_w','finalRating2','finalRating2_w','finalRating3','finalRating3_w']].sort_values(by='finalRating1', ascending  = False).head(10)
```
