```python
inTweet = df[df['countTweet'].notnull()]
inTweet.reset_index(drop=True, inplace=True)

inTweet
```

```python
import jenkspy

inTweet['ratingNoTweet'] = pd.cut(
    inTweet['countTweet'],
    bins=jenkspy.jenks_breaks(inTweet['countTweet'], nb_class=5),
    labels=[1,2,3,4,5],
    include_lowest=True)
inTweet['ratingNoTweet'] = inTweet['ratingNoTweet'].astype('float64')
inTweet['HashReview'] = inTweet.countHashReview/inTweet.countTweet
inTweet['AccountReview'] = inTweet.countAccountReview/inTweet.countTweet

inTweet['ratingHashReview'] = pd.cut(
    inTweet['HashReview'],
    bins=jenkspy.jenks_breaks(inTweet['HashReview'], nb_class=5),
    labels=[1,2,3,4,5],
    include_lowest=True)
inTweet['ratingHashReview'] = inTweet['ratingHashReview'].astype('float64')

inTweet['ratingAccountReview'] = pd.cut(
    inTweet['AccountReview'],
    bins=jenkspy.jenks_breaks(inTweet['AccountReview'], nb_class=5),
    labels=['1', '2' ,'3', '4' ,'5'],
    include_lowest=True)
inTweet['ratingAccountReview'] = inTweet['ratingAccountReview'].astype('float64')

inTweet['ratingTwitter'] = (inTweet.ratingAccountReview+inTweet.ratingHashReview+inTweet.ratingNoTweet)/3
inTweet['ratingTwitter'] = inTweet['ratingTwitter'].round(1)
inTweet = inTweet[['hashtag', 'ratingNoTweet', 'ratingHashReview','ratingAccountReview','ratingTwitter']]
inTweet
```

```python
inTweet.dtypes
```

```python
df = df.join(inTweet.set_index('hashtag'), on='hashtag', how='left')
df['ratingTwitter'] = df['ratingTwitter'].fillna(0)
#df
```

```python
df['ratingCount'] = df['ratingCount'].str.replace(',', '')
df['viewer'] = df['viewer'].str.replace(',', '')
df[['ratingCount', 'viewer']] = df[['ratingCount', 'viewer']].apply(pd.to_numeric)

df['ratingLike'] = pd.cut(
    df['ratingCount'],
    bins=jenkspy.jenks_breaks(df['ratingCount'], nb_class=5),
    labels=['1', '2' ,'3', '4' ,'5'],
    include_lowest=True)


df['ratingView'] = pd.cut(
    df['viewer'],
    bins=jenkspy.jenks_breaks(df['viewer'], nb_class=5),
    labels=['1', '2' ,'3', '4' ,'5'],
    include_lowest=True)

df[['ratingView','ratingLike']] = df[['ratingView','ratingLike']].astype('float64')

df['ratingWebsite'] = (((df.ratingView*2)+(df.ratingLike))/3).round(1)
```

```python
df['finalRating'] = (df.ratingWebsite + (df.ratingTwitter*2))/3 
df['finalRating'] = df['finalRating'].round(1)
df['finalRating'].values[df['finalRating'].values < 1.0] = 1.0
df
```
