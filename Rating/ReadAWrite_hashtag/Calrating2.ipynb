```python
#Calculate rating for twitter
inTweet = df[df['countTweet'].notnull()]
inTweet.reset_index(drop=True, inplace=True)

import jenkspy

inTweet['ratingNoTweet_j'] = pd.cut(
    inTweet['countTweet'],
    bins=jenkspy.jenks_breaks(inTweet['countTweet'], nb_class=5),
    labels=[1,2,3,4,5],
    include_lowest=True)
inTweet['ratingNoTweet_j'] = inTweet['ratingNoTweet_j'].astype('float64')
#inTweet['HashReview'] = inTweet.countHashReview/inTweet.countTweet
#inTweet['AccountReview'] = inTweet.countAccountReview/inTweet.countTweet
inTweet['ratingNoTweet_q'] = pd.qcut(inTweet['countTweet'][inTweet['countTweet'] != 1], 4, labels=[2,3,4,5], duplicates = 'drop')
inTweet['ratingNoTweet_q'] = inTweet['ratingNoTweet_q'].astype('float64')
inTweet['ratingNoTweet_q'] = inTweet['ratingNoTweet_q'].fillna(1)

#Model 3 and 4
inTweet['ratingHashReview_j'] = pd.cut(
    inTweet['countHashReview'],
    bins=jenkspy.jenks_breaks(inTweet['countHashReview'], nb_class=5),
    labels=False,
    include_lowest=True, duplicates='drop')
inTweet['ratingHashReview_j'] = inTweet['ratingHashReview_j'].astype('float64')
inTweet['ratingHashReview_q'] = pd.qcut(inTweet['countHashReview'][inTweet['countHashReview'] != 0], 5, labels=False, duplicates = 'drop')
inTweet['ratingHashReview_q'] = inTweet['ratingHashReview_q'].astype('float64')
inTweet['ratingHashReview_q'] = inTweet['ratingHashReview_q'].fillna(1.7)
inTweet['ratingHashReview_q'] = inTweet['ratingHashReview_q'].replace(0.0,3.3)
inTweet['ratingHashReview_q'] = inTweet['ratingHashReview_q'].replace(1.0,5)

inTweet['ratingAccountReview_j'] = pd.cut(
    inTweet['countAccountReview'],
    bins=jenkspy.jenks_breaks(inTweet['countAccountReview'], nb_class=5),
    labels=False,
    include_lowest=True, duplicates='drop')
inTweet['ratingAccountReview_j'] = inTweet['ratingAccountReview_j'].astype('float64')
inTweet['ratingAccountReview_q'] = pd.qcut(inTweet['countAccountReview'][inTweet['countAccountReview'] != 0], 5, labels=False, duplicates = 'drop')
inTweet['ratingAccountReview_q'] = inTweet['ratingAccountReview_q'].fillna(1.3)
inTweet['ratingAccountReview_q'] = inTweet['ratingAccountReview_q'].replace(0.0,2.5)
inTweet['ratingAccountReview_q'] = inTweet['ratingAccountReview_q'].replace(1.0,3.8)
inTweet['ratingAccountReview_q'] = inTweet['ratingAccountReview_q'].replace(2.0,5.0)

if inTweet['ratingAccountReview_j'].unique().all() < 4.0 and inTweet['ratingAccountReview_j'].unique().all() < 4.0 :
    inTweet[['ratingHashReview_j','ratingAccountReview_j']] = inTweet[['ratingHashReview_j','ratingAccountReview_j']].replace(3.0,5.0) 
    inTweet[['ratingHashReview_j','ratingAccountReview_j']] = inTweet[['ratingHashReview_j','ratingAccountReview_j']].replace(2.0,4.0) 
    inTweet[['ratingHashReview_j','ratingAccountReview_j']] = inTweet[['ratingHashReview_j','ratingAccountReview_j']].replace(1.0,3.0)
    inTweet[['ratingHashReview_j','ratingAccountReview_j']] = inTweet[['ratingHashReview_j','ratingAccountReview_j']].replace(0.0,2.0) 
    

inTweet['ratingTwitter3_j'] = ((inTweet.ratingAccountReview_j+inTweet.ratingHashReview_j+inTweet.ratingNoTweet_j)/3).round(1)
inTweet['ratingTwitter4_j'] = ((inTweet.ratingAccountReview_j+inTweet.ratingHashReview_j+(inTweet.ratingNoTweet_j*2))/4).round(1)
inTweet['ratingTwitter3_q'] = ((inTweet.ratingAccountReview_q+inTweet.ratingHashReview_q+inTweet.ratingNoTweet_q)/3).round(1)
inTweet['ratingTwitter4_q'] = ((inTweet.ratingAccountReview_q+inTweet.ratingHashReview_q+(inTweet.ratingNoTweet_q*2))/4).round(1)
inTweet = inTweet[['hashtag', 'ratingNoTweet_j', 'ratingHashReview_j','ratingAccountReview_j', 'ratingNoTweet_q', 'ratingHashReview_q','ratingAccountReview_q','ratingTwitter3_j','ratingTwitter4_j','ratingTwitter3_q','ratingTwitter4_q']]
inTweet
```

```python
#rating of website
df['ratingCount'] = df['ratingCount'].str.replace(',', '')
df['viewer'] = df['viewer'].str.replace(',', '')
df[['ratingCount', 'viewer']] = df[['ratingCount', 'viewer']].apply(pd.to_numeric)

df['ratingLike_j'] = pd.cut(
    df['ratingCount'],
    bins=jenkspy.jenks_breaks(df['ratingCount'], nb_class=5),
    labels=['1', '2' ,'3', '4' ,'5'],
    include_lowest=True)
df['ratingLike_q'] = pd.qcut(df['ratingCount'], 5, labels=[1,2,3,4,5])

df['ratingView_j'] = pd.cut(
    df['viewer'],
    bins=jenkspy.jenks_breaks(df['viewer'], nb_class=5),
    labels=['1', '2' ,'3', '4' ,'5'],
    include_lowest=True)
df['ratingView_q'] = pd.qcut(df['viewer'], 5, labels=[1,2,3,4,5])

df[['ratingView_j','ratingLike_j','ratingView_q','ratingLike_q']] = df[['ratingView_j','ratingLike_j','ratingView_q','ratingLike_q']].astype('float64')

df['ratingWebsite_j'] = (((df.ratingView_j*2)+(df.ratingLike_j))/3).round(1)
df['ratingWebsite_q'] = (((df.ratingView_q*2)+(df.ratingLike_q))/3).round(1)

df[['title','viewer','ratingCount','ratingView_j','ratingLike_j','ratingView_q','ratingLike_q','ratingWebsite_j','ratingWebsite_q']] 
```
