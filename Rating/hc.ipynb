```python
import pandas as pd
import jenkspy

"""twitter = {
    #'NoTweet': [34,13,11,39,126,28,8,26,5,1,29,0,0,2,0,27,0,0],
    'Re1': [4,0,4,7,1,0,0,5,1,0,1,0,0,0,0,2,0,0]
}"""
#df = pd.DataFrame(twitter)
df = pd.read_csv(r'Twitter_Re_Fav.csv')
df
```

```python
df['RatingRe1'] = pd.cut(
    df['Re2'],
    bins=jenkspy.jenks_breaks(df['Re2'], nb_class=2),
    labels=['1', '2'],
    include_lowest=True)
df
```

```python
df['RatingFav1'] = pd.cut(
    df['Fav2'],
    bins=jenkspy.jenks_breaks(df['Fav2'], nb_class=5),
    labels=['1', '2', '3', '4' , '5'],
    include_lowest=True)
```

```python
df.to_csv(r'Twitter_Re_Fav.csv')
df
```
